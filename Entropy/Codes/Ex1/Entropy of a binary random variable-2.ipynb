# Part 1

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
from math import log2
```

```python
# define the entropy function
def Entropy(probs):
    H = []
    for p in probs:
        if p == 0 or p == 1:
            H.append(0)
            continue
        else:
            h = -p * np.log2(p) - (1 - p) * np.log2(1 - p)
            H.append(h)
    return H
        
```

```python
# define the array range
n = 1000
probs =  np.linspace(0, 1, n)
```

```python
# calculate the entropy for every prob
H = Entropy(probs)
```

```python
# plot the entropy
plt.figure(figsize=(8, 6))
plt.plot(probs, H, label='Entropy', color='g', linewidth=2)
plt.xlabel('Probabilities', fontsize = 14)
plt.ylabel('Entropy of a random binary variable', fontsize = 14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
def ReyniEntropy(alpha, probs):
    ### H(x) = (1/(1-alpha))*log2((p^alpha) + ((1-p)^alpha))
    H = []
    for p in probs:
        if p == 0 or p == 1:
            H.append(0)
            continue
        else:
            h = (1/(1-alpha)) * (np.log2((np.power(p,alpha)) + np.power((1 - p),alpha)))
            H.append(h)
    return H
```

```python
# define the array range
n = 1000
probs =  np.linspace(0, 1, n)
```

```python
# calculate the Reyni entropy for every prob for alpha = 0.5
alpha = 0.5
H = ReyniEntropy(alpha,probs)
```

```python
# plot the entropy
plt.figure(figsize=(8, 6))
plt.plot(probs, H, label='Entropy for alpha = 0.5', color='b', linewidth=2)
plt.xlabel('Probabilities', fontsize = 14)
plt.ylabel('ReyniEntropy', fontsize = 14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
